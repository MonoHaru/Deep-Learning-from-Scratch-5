```python
import numpy as np
```

```python
x = np.array([1, 2, 3])

print(x.__class__)
print(x.shape)
print(x.ndim)
```

```python
W = np.array([[1, 2, 3],
              [4, 5, 6]])
print(W.ndim)
print(W.shape)
```

```python
W = np.array([[1, 2, 3], [4, 5, 6]])
X = np.array([[0, 1, 2], [3, 4, 5]])

print(W + X)
print('---')
print(W * X)  # 아다마르 곱(Hadamard product)
```

```python
# 벡터의 내적
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
y = np.dot(a, b)
print(y)

# 행렬곱
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
Y = np.dot(A, B)
print(Y)
```

```python
# 벡터의 내적
y = a @ b
print(y)

# 행렬곱
Y = A @ B
print(Y)
```
