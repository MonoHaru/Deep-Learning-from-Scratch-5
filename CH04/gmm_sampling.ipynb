```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ========== 학습된 매개변수 ==========
mus = np.array([[2.0, 54.50],
                [4.3, 80.0]])
covs = np.array([[[0.07, 0.44],
                  [0.44, 33.7]],
                 [[0.17, 0.94],
                  [0.94, 36.00]]])
phis = np.array([0.35, 0.65])
# ====================================
```

```python
def sample():
  z = np.random.choice(2, p=phis)
  mu, cov = mus[z], covs[z]
  x = np.random.multivariate_normal(mu, cov)
  return x
```

```python
N = 500  # 샘플 개수
xs = np.zeros((N, 2))
for i in range(N):
  xs[i] = sample()  # 샘플 추출
```

```python
plt.scatter(xs[:, 0], xs[:, 1], color='orange', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
