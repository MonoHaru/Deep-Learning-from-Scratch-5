```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_sums = []
N = 5
```

```python
for _ in range(10000):
  xs = []
  for n in range(N):
    x = np.random.rand()  # 균등 분포로부터 무작위 수 추출
    xs.append(x)
  t = np.sum(xs)  # 표본 합 계산
  x_sums.append(t)
```

```python
# 정규 분포
def normal(x, mu=0, sigma=1):
  y = 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-(x - mu)**2 / (2 * sigma**2))
  return y
```

```python
x_norm = np.linspace(-5, 5, 1000)
mu = N / 2
sigma = np.sqrt(N / 12)
y_norm = normal(x_norm, mu, sigma)
```

```python
plt.hist(x_sums, bins='auto', density=True)
plt.plot(x_norm, y_norm)
plt.title(f'N={N}')
plt.xlim(-1, 6)  # x축의 범위를 -1 ~ 6으로 설정
plt.show()
```
