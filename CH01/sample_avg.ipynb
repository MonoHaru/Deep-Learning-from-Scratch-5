```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 3  # 샘플 크기

xs = []
for n in range(N):
  x = np.random.rand()  # 균등 분포로부터 무작위 샘플링
  xs.append(x)

x_mean = np.mean(xs)
print(x_mean)
```

```python
x_means = []
N = 1  # 우선 샘플 크기가 1인 경우부터

for _ in range(1000):
  xs = []
  for n in range(N):
    x = np.random.rand()  # 균등 분포로부터 샘플링
    xs.append(x)
  mean = np.mean(xs)  # 표본 평균
  x_means.append(mean)

# 그래프 그리기
plt.hist(x_means, bins='auto', density=True)
plt.title(f'N = {N}')
plt.xlabel('x')
plt.ylabel('Probability Density')  # y축은 '확률 밀도'
plt.xlim(-0.05, 1.05)  # x축 범위
plt.ylim(0, 5)  # y축 범위
plt.show()
```
