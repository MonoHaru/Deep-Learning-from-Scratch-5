```python
import numpy as np
from scipy.stats import norm
```

```python
path = '/content/drive/MyDrive/DLFS5/CH02/height.txt'
xs = np.loadtxt(path)
mu = np.mean(xs)
sigma = np.std(xs)
```

```python
p1 = norm.cdf(160, mu, sigma)
print('p(x <= 160):', p1)  # 키가 160cm 이하일 확률
```

```python
p2 = norm.cdf(180, mu, sigma)
print('p(x > 180):', 1-p2)  # 키가 180cm보다 클 확률
```
