```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "/content/drive/MyDrive/DLFS5/CH02/height.txt"
xs = np.loadtxt(path)
print(xs.shape)
```

```python
plt.hist(xs, bins='auto', density=True)
plt.xlabel('Height(cm)')
plt.ylabel('Probability Density')
plt.show()
```

```python
# xs눈 25,000개의 원소로 이루어진 1차원 배열
mu = np.mean(xs)
sigma = np.std(xs)

print(mu)
print(sigma)
```
