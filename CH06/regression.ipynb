```python
import torch
```

```python
# 토이 데이터셋
torch.manual_seed(0)  # 시드 고정
x = torch.rand(100, 1)
y = 2 * x + 5 + torch.rand(100, 1)
```

```python
W = torch.zeros((1, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

```python
def predict(x):
  y = x @ W + b
  return y
```

```python
def mean_squared_error(x0, x1):  # 평균 제곱 오차 계산
  diff = x0 - x1
  N = len(diff)
  return torch.sum(diff ** 2) / N
```

```python
lr = 0.1     # 학습률
iters = 100  # 반복 횟수
```

```python
for i in range(iters):  # 경사법
  y_hat = predict(x)
  loss = mean_squared_error(y, y_hat)

  loss.backward()

  W.data -= lr * W.grad.data
  b.data -= lr * b.grad.data

  W.grad.zero_()
  b.grad.zero_()

  if i % 10 == 0:  # 10회마다 출력
    print(loss.item())

print(loss.item())
print('====')
print('W =', W.item())
print('b =', b.item())
```
